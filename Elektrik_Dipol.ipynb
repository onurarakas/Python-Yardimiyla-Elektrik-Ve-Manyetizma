```python
import matplotlib.pyplot as plt
import numpy as np

#!ELektrik Dipol izimi iin bir kod


#axis ayar
N = 40
M = 40
#koordinatlar
X = np.arange(0, M, 1)
Y = np.arange(0, N, 1)
X, Y = np.meshgrid(X, Y)

Ex ,Ey = np.zeros((N,M)), np.zeros((N,M)) #Elektrik alanmn x ve y si iin depo 

konum = [[], []] 

konum_x = 25
konum_y = 20

konum[0].append(konum_x)
konum[1].append(konum_y)

yuk = (11)*(10**(-19)) 

for i in range(N):
    for j in range(M):
        payda = ((i - konum_y)**2 + (j - konum_x)**2)**1.5  
        if payda != 0:
             Ex[i, j] += yuk*(8.99*10**9)*(j - konum_x) / payda
             Ey[i, j] += yuk*(8.99*10**9)*(i - konum_y) / payda



Ex1 ,Ey1 = np.zeros((N,M)), np.zeros((N,M))

konum1 = [[], []] 

konum_x1 = 20 
konum_y1 = 20 

konum1[0].append(konum_x1)
konum1[1].append(konum_y1)

yuk1 = (-11)*(10**(-19)) 

for i in range(N):
    for j in range(M):
        payda = ((i - konum_y1)**2 + (j - konum_x1)**2)**1.5  
        if payda != 0:
             Ex1[i, j] += yuk1*(8.99*10**9)*(j - konum_x1) / payda
             Ey1[i, j] += yuk1*(8.99*10**9)*(i - konum_y1) / payda

plt.figure(figsize=(12, 8))

EX =Ex + Ex1 #iki farkl datay tek bir yere alabilmek iin onlar topluyorum
EY =Ey + Ey1

# izim
#plt.plot(*konum, 'bo')
fig, ax = plt.subplots(figsize = (12,10))
q = ax.streamplot(X, Y, EX, EY)
ax.add_artist(plt.Circle((konum_x,konum_y), 0.17, color='r'))
ax.add_artist(plt.Circle((konum_x1,konum_y1), 0.17, color='b'))
#plt.streamplot(X, Y, Ex, Ey)
#plt.streamplot(X, Y, Ex1, Ey1)
plt.title('Elektirk Alan')
plt.axis('equal')
plt.show()
```
